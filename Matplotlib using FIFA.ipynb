```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv(r'C:\Users\USER\Desktop\DataSet\fifa_data.csv')
```

# Histograms

```python
bins=[40,50,60,70,80,90,100]
plt.hist(data.Overall,bins=bins,color='#abcdef')
plt.xticks(bins)

plt.ylabel('Number of players ')
plt.xlabel('Skill Level')
plt.title('Distribution of players skills in FIFA 2018',)
plt.show()
```

# Pie CHarts

```python
left=data.loc[data['Preferred Foot']=='Left'].count()[0]
right=data.loc[data['Preferred Foot']=='Right'].count()[0]
print(right)
print(left)
```

```python
labels=['Left','Right']
plt.pie([left,right],labels=labels,autopct='%.2f%%')

plt.title('Foot Preference of FIFA players')
plt.show()
```

```python
data.Weight
```

```python
data.Weight=[int(x.strip('lbs') )if type(x)==str else x for x in data.Weight]

plt.style.use('ggplot')

light = data.loc[data.Weight < 125].count()[0]
light_medium = data.loc[(data.Weight >= 125) & (data.Weight < 150)].count()[0]
medium = data[(data.Weight >= 150) & (data.Weight < 175)].count()[0]
medium_heavy = data[(data.Weight >= 175) & (data.Weight < 200)].count()[0]
heavy = data[(data.Weight >= 200)].count()[0]
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-150','150-175','175-200','Over 200']
explode=[.4,.2,0,0,.4]
plt.pie(weights,labels=labels,explode=explode)
plt.title('Weight Distribution of FIFA Player')
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5,8))
barcelona = data.loc[data.Club=='FC Barcelona']['Overall']
madrid=  data.loc[data.Club=='Real Madrid']['Overall']
DSA = data.loc[data.Club == 'New England Revolution']['Overall']

labels=['FC Barcelona','Real Madrid','New England Revolution']
boxes=plt.boxplot([barcelona,madrid,DSA],labels=labels,patch_artist=True,medianprops={'linewidth':2})

for box in boxes['boxes']:
    #set edge colors
    box.set(color='#abcdef',linewidth=2)

    #change fill color
    
    box.set(facecolor='#27383D')
plt.title('Professonal Soccer Team ')
plt.show()
```
